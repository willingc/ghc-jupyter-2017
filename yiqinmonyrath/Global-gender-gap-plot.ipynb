```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Loading the dataset to a Pandas dataframe

```python
global_gender = pd.read_excel("global-gender-gap-index-2013.xls", encoding="utf-8", index_col=False)
global_gender
```

### Reading rank and score for top 5 countries from 2006 to 2013

```python
rank = []
year = []
score = []
templist = []
templist1 = []
for x in range(0,8):
    temp = 2006 + x
    colName = str(temp) + " Rank"
    colNameScore = str(temp) + " Score"
    year.append(temp)
    for c in range(0,5):
        templist.append(int(global_gender.at[c,colName]))
        templist1.append(global_gender.at[c,colNameScore])
    rank.append(templist)
    score.append(templist1)
    templist1 = []
    templist = []
```

### Plot rank and score

```python
dfrank = pd.DataFrame(rank,year,columns=['Iceland','Finland','Norway','Sweden','Philippines'])
dfscore = pd.DataFrame(score,year,columns=['Iceland','Finland','Norway','Sweden','Philippines'])
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,6))


resultRank = dfrank.plot(title='Top 5 countries rank from 2006 to 2013', ax=axes[0])
resultRank.set_xlabel('Year')
resultRank.set_ylabel('Rank')

resultScore = dfscore.plot(title='Top 5 countries score from 2006 to 2013', ax=axes[1])
resultScore.set_xlabel('Year')
resultScore.set_ylabel('Score')

fig.tight_layout()
```
