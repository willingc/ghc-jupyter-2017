```python
%matplotlib inline

import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")

from matplotlib import pyplot as plt
```

```python
infantMortality = pd.read_csv("OSD_BDChildData.csv")
```

```python
infantMortality
```

```python
cs=sns.color_palette("cubehelix", 10)
```

```python
import pprint as pp
def plotdat(l, title_text='Infant mortality rate in Bangladesh', sortON=True, plot_type="line"):
    if sortON == True:
        l.sort_values(inplace=True)
    fig=plt.figure(figsize=(10,5))
    plt.yticks(fontsize=8)
    
    
    if plot_type=="line":
        l.plot(kind=plot_type, fontsize=12,color = cs[1], marker='o')
    else:
        l.plot(kind=plot_type, fontsize=12,color = cs[1])

    
    plt.xlabel('')
    plt.ylabel('Infant mortality rate',fontsize=10)
    plt.title(title_text, fontsize=20)
```

This data shows the infant mortality rate in bangladesh. In the graph, we see that the mortality rate has decreased in a linear fashion over the time.
Bangladesh is a developing country with a large population and lots of resource problems. But for past few years it has
been really trying to become rich in one thing, which is technology. We call this initiative as "Digital Bangladesh". The graph shows one of the result of that initative.
We have many mobile apps now which increases health awareness for new mothers and infants. Mothers are much more educated about the birth process now than before. What are the dos and donts at what stage, what to eat, how to take care of the infant etc. This result is exactly reflecting what we have achieved in last one decade.
